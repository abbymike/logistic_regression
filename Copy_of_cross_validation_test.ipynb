```python
logreg=LogisticRegression()
```

```python
x_train,x_test,y_train,y_test=train_test_split(data,tr,test_size=0.3,random_state=42)
logreg.fit(x_train,y_train)
predict=logreg.predict(x_test) #Train-Test Split
```

```python
print("Accuracy score on training set is {}".format(accuracy_score(logreg.predict(x_train),y_train)))
print("Accuracy score on test set is {}".format(accuracy_score(predict,y_test)))
```

```python
kf=KFold(n_splits=5)
kfold = cross_val_score(logreg,data,tr,cv=kf)
print("Cross Validation Scores are {}".format(kfold))
print("Average Cross Validation score :{}".format(kfold.mean()))
kfolds = kfold.mean()
print(kfolds) #kfold

stratifiedkf=StratifiedKFold(n_splits=5)
stratifiedkfold = cross_val_score(logreg,data,tr,cv=stratifiedkf)
print("Cross Validation Scores are {}".format(stratifiedkfold))
print("Average Cross Validation score :{}".format(stratifiedkfold.mean()))
stratifiedkfolds = (stratifiedkfold.mean())
print(stratifiedkfolds) #stratifiedkfolds
```
